```python

import os
from utils import config
import numpy as np
import matplotlib.pyplot as plt

```

```python
date = "2019-03-18 10:08:58"
filepath = os.path.abspath(os.path.join(config().sim.output.path, date))

print(filepath)
```

```python
mean = np.load(filepath+"/returns_intra.npy").mean(1).squeeze()
std = np.load(filepath+"/returns_intra.npy").std(1).squeeze()
t = range(0, len(mean))
plt.figure()
plt.plot(t, mean)
plt.fill_between(t, mean-std, mean+std, alpha=0.5, edgecolor='#608cd1', facecolor='#bed3f4')
plt.xlabel("Episodes")
plt.ylabel("Returns intrasec")
plt.title(f"Agent")
plt.savefig(f"returns_agent.eps", type="eps", dpi=1000)
plt.show()
```

```python
mean = np.load(filepath+"/losses_forward.npy").squeeze()
t = range(0, len(mean))
plt.figure()
plt.plot(t, mean)
#plt.fill_between(t, mean, mean, alpha=0.5, edgecolor='#608cd1', facecolor='#bed3f4')
plt.xlabel("Episodes")
plt.ylabel("Loss forward")
plt.title(f"Agent")
plt.savefig(f"returns_agent.eps", type="eps", dpi=1000)
plt.show()
```

```python
returns_all = np.load(filepath+"/returns.npy")
print(returns_all.shape)
mean = returns_all.mean(1)
std = returns_all.std(1)
print(mean.shape)
t = range(0, len(mean))
plt.figure()
plt.plot(t, mean)
plt.fill_between(t, mean-std, mean+std, alpha=0.5, edgecolor='#608cd1', facecolor='#bed3f4')
plt.xlabel("Episodes")
plt.ylabel("Return al")
plt.title(f"Agent")
plt.savefig(f"returns_agent.eps", type="eps", dpi=1000)
plt.show()
```
